```python
flight_data['order_date'] = pd.to_datetime(flight_data.order_date)
flight_data['deliver_date'] = pd.to_datetime(flight_data['deliver_date'])
flight_data['planned_date'] = pd.to_datetime(flight_data['planned_date'])
flight_data.set_index('order_date', inplace=True) # after setting order_date as an index, it will not be in the column
df = flight_data.loc['2019-3-21':'2019-3-28']  # order date in the range
```

```python
df.reset_index(level=0, inplace=True) # need to reset the index so that the order_date will be back in column
df.sort_values('order_date', inplace=True, ascending=False)
df.reset_index(level=0, inplace=True) # need to reset the index AGAIN so that the orderrange list would be the same order
```

```python
df['from_to'] = df['origin'] + ' - '+ df['desti']
```

```python
df.loc[:,'color'] = np.where((df['risk_all'] >= 6) & (df['value'] >= 30000), 'red', 
                            np.where((df['risk_all'] <= 3) & (df['value'] <= 60000), 'green', 'orange'))
```

```python
from bokeh.io import show, output_file, output_notebook
from bokeh.models import ColumnDataSource, SingleIntervalTicker, LinearAxis, LinearColorMapper, ColorBar
from bokeh.plotting import figure
from bokeh.models.tools import HoverTool
from bokeh.transform import transform

#output_file("iris.html", title="iris.py example")
output_notebook

source = ColumnDataSource(df)

risk_color_mapper = LinearColorMapper(palette=['#2BA743', '#F5A623', '#E90F0F'], 
                                 low=0, high=10)
value_color_mapper = LinearColorMapper(palette=['#2BA743', '#F5A623', '#E90F0F'], 
                                 low=0, high=100000)

p = figure(title = "Risk Value Chart", plot_width=800, plot_height=600,toolbar_location=None, 
           x_range=(0, 10), y_range=(0, 100000))
p.xaxis.ticker = SingleIntervalTicker(interval=1, num_minor_ticks=0)
p.xaxis.axis_label = 'Risk'
p.yaxis.axis_label = 'Value'
# p.xaxis.axis_line_width = 3
# p.xaxis.bounds = (7, 10)
# p.xaxis.axis_line_color = "red"

p.circle('risk_all','value',source=source, color='color',
         alpha=0.5, size=10)  # alpha and fill_alpha are different!

# p.scatter('risk_all','value',source=source, radius=0.05, 
#           fill_color='blue', fill_alpha=0.6,
#           line_color=None)

color_bar = ColorBar(color_mapper=risk_color_mapper, location=(0,0), orientation='horizontal')
p.add_layout(color_bar, 'below')
color_bar = ColorBar(color_mapper=value_color_mapper, location=(0,0), orientation='vertical')
p.add_layout(color_bar, 'left') 
# how to make colorbar scientific false??

p.left[0].formatter.use_scientific = False 

hover = HoverTool()
hover.tooltips = [
            ('Order number','#@i1_legid'),
            ('From to', '@from_to'),
            ('Risk', '@risk_all'),
            ('Value', '@value'),
            ('Legs', '@i1_hops'),
            ('Expected delivery time', '@deliver_date{%F}'),
           ]

hover.mode = 'mouse'
hover.formatters={'deliver_date': 'datetime'}

p.add_tools(hover)

show(p)
```
