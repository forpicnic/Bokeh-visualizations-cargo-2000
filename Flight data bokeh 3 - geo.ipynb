```python
import numpy as np
import pandas as pd

flight_data = pd.read_csv('/Users/xuxinyi/Downloads/sort.csv', index_col=0) # remove the index column
flight_data.head()
```

```python
# df.drop('column_name', axis=1, inplace=True)  # To delete the column without having to reassign df
flight_data.drop(['i1_rcf_1_p', 'i1_rcf_1_e', 'i1_rcf_2_p', 'i1_rcf_2_e', 'i1_rcf_3_p', 'i1_rcf_3_e'], axis=1, inplace=True)
```

```python
flight_data['from_to'] = flight_data['origin'] + ' - '+ flight_data['desti']
```

```python
path = '/Users/xuxinyi/Downloads/d3/data'
airport_nodes = pd.read_csv(path + '/flight-data/air-routes-latest-nodes.csv')
airport_edges = pd.read_csv(path + '/flight-data/air-routes-latest-edges.csv')

take_rows = ['FRA', 'LHR', 'CDG', 'AMS', 
             'FCO', 'MUC', 'ZRH', 'BLA',
             'MAD', 'ARN', 'DUB', 'CPH', 
             'CRL', 'VIE', 'OSL',
             'PVG', 'DXB', 'DOH', 'ORD',
             'SYD', 'MEM', 'SIN', 'HKG',
             'SVO', 'DEL', 'SDF', 'IST',
             'BKK', 'TLV']


nodes = airport_nodes.loc[airport_nodes['code'].isin(take_rows)]
nodes.head()
```

```python
# join two datasets, 
# sort.csv doesn't have 3 hops..In the original dataset, 23 shippments has 3 hops among 3760 shippments..
mapping_lat = dict(zip(nodes.code, nodes.lat))
mapping_lon = dict(zip(nodes.code, nodes.lon))
mapping_lat.update({'?': None})
mapping_lon.update({'?': None})

flight_data['i1_dep_1_place_x'] = flight_data['i1_dep_1_place'].replace(mapping_lat)
flight_data['i1_dep_1_place_y'] = flight_data['i1_dep_1_place'].replace(mapping_lon)
flight_data['i1_rcf_1_place_x'] = flight_data['i1_rcf_1_place'].replace(mapping_lat)
flight_data['i1_rcf_1_place_y'] = flight_data['i1_rcf_1_place'].replace(mapping_lon)
flight_data['i1_rcf_2_place_x'] = flight_data['i1_rcf_2_place'].replace(mapping_lat)
flight_data['i1_rcf_2_place_y'] = flight_data['i1_rcf_2_place'].replace(mapping_lon)
```

```python
from bokeh.io import output_file, show, output_notebook
from bokeh.models import ColumnDataSource, GMapOptions, Arrow, NormalHead
from bokeh.models.tools import HoverTool
from bokeh.plotting import gmap

#output_file("gmap.html")
output_notebook()

place_x = flight_data['place_x'].tolist()
place_y = flight_data['place_y'].tolist()
source = ColumnDataSource(flight_data)

#map_options = GMapOptions(lat=30.2861, lng=-97.7394, map_type="roadmap", zoom=2)
map_options = GMapOptions(map_type="roadmap", zoom=2)

# For GMaps to function, Google requires you obtain and enable an API key:
#
#     https://developers.google.com/maps/documentation/javascript/get-api-key

p = gmap("AIzaSyBEXeQwbMgs0zCkvYsKcES_A6pvePsehtw", map_options, title="Route Map", 
         plot_height=600, plot_width=1000, tools='hover,pan,wheel_zoom')  # zoom_in,zoom_out not working

p.circle(x="i1_dep_1_place_y", y="i1_dep_1_place_x", size=5, fill_color="blue", fill_alpha=1, source=source)
p.circle(x="i1_rcf_1_place_y", y="i1_rcf_1_place_x", size=5, fill_color="blue", fill_alpha=1, source=source)
p.circle(x="i1_rcf_2_place_y", y="i1_rcf_2_place_x", size=5, fill_color="blue", fill_alpha=1, source=source)

# p.multi_line(place_y, place_x,
#              color="firebrick", alpha=0.3, line_width=0.5,
#              hover_color = 'blue', hover_alpha = 1.0)  # why google map the X,Y are opposite?

p.multi_line("place_y", "place_x",source=source,
             color="firebrick", alpha=0.3, line_width=0.5,
             hover_color = 'blue', hover_alpha = 1.0)  # why google map the X,Y are opposite?

# if hover is added after multi_line, then hover over multi_line; if hover is added after circles, then hover over circles
hover = HoverTool()
hover.tooltips = [
            ('Order number','#@i1_legid'),
            ('From to','@from_to'),
            ('Expected delivery time', '@deliver_date{%F}'),
            ('Legs','@i1_hops'), 
            ('Overall risk','@risk_all{0.0}'),
            ('Leg 1 risk', '@risk_1{0.0}'),
            ('Leg 2 risk', '@risk_2{0.0}')
           ]

hover.mode = 'mouse'
hover.formatters={'deliver_date': 'datetime'}
p.add_tools(hover)

show(p)
```
